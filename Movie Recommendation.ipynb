# Data Collection and Pre-Processing

```python
data = pd.read_csv('movies.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
# Selceting the relevent features for recommendation
selected_feature = ['genres','keywords','cast','tagline','director']
print(selected_feature)
```

```python
# replacing the missing values or null values with null string
for feature in selected_feature:
    data[feature] = data[feature].fillna('')
```

```python
# cobining all the 5 selected features
combined_feature = data['genres'] + data['keywords'] + data['cast'] + data['tagline'] + data['director']
```

```python
combined_feature
```

```python
# Converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

# Cosine Similarity

```python
# Getting similarity score  using cosine similarity
similarity = cosine_similarity(feature_vector)
print(similarity)
```

```python
similarity.shape
```

```python
# Getting the Movie name from the user
movie_name = input("Enter your favourite movie name : ")
```

```python
# creating the list all the movie name given in the dataset
list_of_titles =  data['title'].tolist()
print(list_of_titles)
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name,list_of_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the Index of the movie with the title
index_of_the_movies = data[data.title==close_match]['index'].values[0]
print(index_of_the_movies)
```

```python
# getting list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movies]))
print(similarity_score)
```

```python
# sorting the movies based on their similarity values
sorted_similar_movies = sorted(similarity_score,key = lambda x :x[1],reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar mvies based on the index
print('Movies suggested for you : \n')
i=1
for movies in sorted_similar_movies:
    index= movies[0]
    title_from_index = data[data.index==index]['title'].values[0]
    if (i<31):
        print(i, '.',title_from_index)
        i+=1
        
```

## Movie Recommendation system

```python
# Getting the Movie name from the user
movie_name = input("Enter your favourite movie name : ")

list_of_titles =  data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name,list_of_titles)

close_match = find_close_match[0]

index_of_the_movies = data[data.title==close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movies]))

sorted_similar_movies = sorted(similarity_score,key = lambda x :x[1],reverse = True)

# print the name of similar mvies based on the index
print('Movies suggested for you : \n')
i=1
for movies in sorted_similar_movies:
    index= movies[0]
    title_from_index = data[data.index==index]['title'].values[0]
    if (i<31):
        print(i, '.',title_from_index)
        i+=1
        
```
